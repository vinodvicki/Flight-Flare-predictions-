## EDA ##

here we have to change the date time objects in to multiple virables by feature engineering

```python
# converting date time objects
train_df['Date_of_Journey']= pd.to_datetime(train_df['Date_of_Journey'], format='%d/%m/%Y')
```

```python
train_df.head(2)
```

```python
plt.figure(figsize=(15,5))
# below plot shows the date wise price graph
train_df.groupby('Date_of_Journey')['Price'].mean().plot(kind='line')
plt.xlabel('Date_of_Journey')
plt.ylabel('Price')
plt.title('Date_of_Journey vs Price')
plt.show()
```

```python
# extractin features from Date_of_Journey column as day, month, year
train_df['day'] = train_df['Date_of_Journey'].dt.day
train_df['month'] = train_df['Date_of_Journey'].dt.month
train_df['year'] = train_df['Date_of_Journey'].dt.year
```

```python
train_df.head(2)
```

```python
train_df.tail(2)
```

```python
# extracting hour and min from arrival_time to know exact time  from arrival_time
train_df['Arrival_Hour'] = pd.to_datetime(train_df['Arrival_Time']).dt.hour  # later we drop Arrival comlumn
train_df['Arrival_Minute'] = pd.to_datetime(train_df['Arrival_Time']).dt.minute
```

```python
train_df.head(2)
```

```python
# extracting hour and min from arrival_time to know exact time  from Dep_Time
train_df['dep_hour'] = pd.to_datetime(train_df['Dep_Time']).dt.hour
train_df['dep_minute'] = pd.to_datetime(train_df['Dep_Time']).dt.minute
train_df.head(2)
```

```python
train_df.Total_Stops.value_counts()
```

```python
train_df['Total_Stops'] = train_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4})
```

```python
#the above cell is coverting the no of stops in to numeric virables
train_df.Total_Stops.value_counts()
```

```python
# Creates a new column duration_hour that contains only the hour part of the Duration column, with any unit or minute part removed.
train_df['duration_hour'] = train_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
train_df['duration_minute'] = train_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
train_df.head(2)
```
